# Importing Libararies

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("/content/wine.csv")
```

```python
df
```

```python
df1=df.copy()
```

```python
df=df.iloc[:,1:]
df
```

```python
df.shape
```

```python
df.info()
```

```python
sc=StandardScaler()
df=sc.fit_transform(df)
```

```python
df
```

# PCA Implementation

```python
pca=PCA(n_components=2)
```

```python
pca.fit_transform(df)
```

```python
p_components=pd.DataFrame(pca.fit_transform(df),columns=['pc1','pc2'])
```

```python
p_components
```

```python
#building PCA for all components
pca1=PCA(n_components=13)
pca1_values=pca1.fit_transform(df)
```

```python
pca1_values
```

```python
#variance of pcs
var = pca1.explained_variance_ratio_
var
```

```python
#cumulative variance
var1 = np.cumsum(var*100)
var1
```

```python
#weights
pca1.components_
```

```python
plt.plot(var1)
plt.show()
```

```python
final_df=pd.concat([df1['Type'],pd.DataFrame(pca1_values[:,0:3],columns=['pc1','pc2','pc3'])],axis=1)
final_df
```

```python
#Visulaisation of PCAs
plt.figure(figsize=(12,10))
sns.scatterplot(data=final_df)
```

# Hierarchial Clustering

```python
#import libararies
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(12,10))
dendogram=sch.dendrogram(sch.linkage(df,'complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
yhc=hc.fit_predict(df)
```

```python
yhc
```

```python
df2=df1.copy()
```

```python
df2['clusterid']=hc.labels_
```

```python
df2
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df2['clusterid'],df2['Type'],c=hc.labels_)
```

# Kmeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,11),wcss)
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit_predict(df)
```

```python
df3=df1.copy()
```

```python
df3['clusterid']=kmeans.labels_
```

```python
df3
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df3['clusterid'],df3['Type'],c=kmeans.labels_)
```
